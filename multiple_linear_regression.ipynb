## Encoding cyclical datetime features

Sources:

https://www.kaggle.com/code/avanwyk/encoding-cyclical-features-for-deep-learning
https://towardsdatascience.com/cyclical-features-encoding-its-about-time-ce23581845ca
https://developer.nvidia.com/blog/three-approaches-to-encoding-time-information-as-features-for-ml-models/

Method:

* encode cyclical data to transform the data into two dimensions using sinus and cosine
* the idea behind using trigonometric functions is that they can represent cyclical patterns by mapping the values to a circle. 

```python
# create sample data 

sample = X[:168]
```

```python
# plot the PU_hour of the sample 

ax = sample['PU_hour'].plot()
```

```python
# Transformation of hourly data

X = X.copy()
X['PU_hour_sin'] = np.sin(2* np.pi * X['PU_hour'] / 24.0)
X['PU_hour_cos'] = np.cos(2* np.pi * X['PU_hour'] / 24.0)
```

```python
X.head()
```

```python
# Plot the sinus and cosinus values

fig, ax = plt.subplots()
ax.scatter(X['PU_hour_sin'], X['PU_hour_cos'])
ax.set_aspect('equal')
ax.set_xlabel('PU_hour_sin')
ax.set_ylabel('PU_hour_cos')
ax.set_title('Sinus and Cosinus values')
plt.show()
```

```python
def encode(data, col, max_val):
    data[col + '_sin'] = np.sin(2 * np.pi * data[col]/max_val)
    data[col + '_cos'] = np.cos(2 * np.pi * data[col]/max_val)
    return data

X = encode(X, 'PU_month', 6)
X = encode(X, 'PU_day_of_month', 31)
X = encode(X, 'PU_day_of_week', 7)
```

In the following code, it will be evaluated how well the linear regression model fits the data, both on the training set and the test set. 
The mean squared error (=MSE) and the root mean squared error (=RMSE) are measures of how far the model's predictions are from the actual values, while the R-squared metric measures how much of the variation in the dependent variable is explained by the independent variables of the model.

```python
# Get the intercept and sloce

print("Intercept: \n", lin_reg_a.named_steps['linearregression'].intercept_)
print("Coefficients: \n", lin_reg_a.named_steps['linearregression'].coef_)

# Evaluate the results on the test data

# Calculate the mean squared error (=MSE), the Root mean squared error (=RMSE) and coefficient of determination (R^2)
mse_a_test = mean_squared_error(y_a_test, y_a_pred_test)
rmse_a_test = np.sqrt(mse_a_test)
r2_a_test = r2_score(y_a_test, y_a_pred_test)

print("Mean squared error: {:.2f}".format(mse_a_test))
print("Root mean squared error: {:.2f}".format(rmse_a_test))
print("Coefficient of determination (R^2): {:.2f}".format(r2_a_test))

# Evaluate the results on the train data 

mse_a_train = mean_squared_error(y_a_train, y_a_pred_train)
rmse_a_train = np.sqrt(mse_a_train)
r2_a_train = r2_score(y_a_train, y_a_pred_train)

print("Mean squared error: {:.2f}".format(mse_a_train))
print("Root mean squared error: {:.2f}".format(rmse_a_train))
print("Coefficient of determination (R^2): {:.2f}".format(r2_a_train))
```

**Coefficients:** <br>
The coefficients shown are the values that were calculated for the parameters of the multiple linear regression model. <br>
These coefficients represent the amount of change in the dependent variable that is associated with a one-unit change in the corresponding independent variable, holding all other independent variables constant. <br>
For example, when variable 2 increases by one unit, the dependent variable is expected to increase by 0.9231, holding all other independent variables constanct. <br>

**MSE:** <br>
The MSE is a measure of how well the model fits the data. <br>
The lower the MSE, the better the model is at predicting the values of the dependent variable. <br>

**RMSE:** <br>
The RMSE is the square root of MSE and provides an estimate of the standard deviation of the errors. In this case, the RMSE on the test data is 20.88 and on the train data is 21.15 and the pickup count is not close to each other. Specifically, this indicates that on average, the predictions of the model are off by around 20-21 pickups.

**R_squared:** <br>
The coefficient of determination (R-squared value) of 0.92 represents the proportion of the variance in the dependent variable that can be explained by the independent variables in the model. <br>
The R-squared value ranges from 0 to 1, where 0 indicates that the model does not explain any of the variance in the dependent variable, and 1 indicates that the model explains all of the variance in the dependent variable. <br>
In this case, the R-squared value of 0.92 indicates that the model explains 92% of the variance in the dependent variable, which is relatively high. <br>
This means that the independent variables used in the model are able to explain 92 % of the variation observed in the dependent variable. <br>
However, it is important to keep in mind that R_squared does not tell us anything about the validity or reliability of the model. Other evaluation metrics (e.g. MSE and RMSE) should be considered to determine the accuracy and quality of the model.

### k-fold cross validation

The following code performs cross-validation on a linear regression model using the method *cross_val_score* from scikit-learn's module. <br>
In this case, the *cv=10* argument specifies that the data will be split into 10 equally sized folds, and the model will be trained and tested on each fold. The *scoring* argument specifies the metric used to evaluate the performance of the model, which is the negative mean squared error. 
Then, the root mean squared error for each fold is calculated and stored in the *rmse_scores_a* array. Finally, the *display_scores* function is defined, which displays the individual scores (10 in total), the mean score, and the standard deviation of the scores.

## Model with MinMaxScaler

**sklearn.preprocessing.MinMaxScaler:** <br>
* This estimator scales and translates each feature individually such that it is in the given range on the training set, e.g. between zero and one 
* The transformation is given by 
     * X_std = (X - X.min(axis=0)) / (X.max(axis=0) - X.min(axis=0))
     * X_scaled = X_std * (max - min) + min
     * where min, may = feature range <br>
* Parameters, e.g. feature_range: tuple(min, max), default = (0,1)

**Source:** <br>
* https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.MinMaxScaler.html#sklearn.preprocessing.MinMaxScaler

# Feature Selection using RFE (=Recursive Feature Elimination)

Given an external estimator that assigns weights to features (e.g., the coefficients of a linear model), the goal of recursive feature elimination (RFE) is to select features by recursively considering smaller and smaller sets of features. First, the estimator is trained on the initial set of features and the importance of each feature is obtained either through any specific attribute or callable. Then, the least important features are pruned from current set of features. That procedure is recursively repeated on the pruned set until the desired number of features to select is eventually reached.


Source: <br>
* https://medium.com/@feraguilari/multiple-regression-and-recursive-feature-elimination-rfe-34af0c6ae51b
* https://www.kaggle.com/code/kavisekar/features-selection-in-multiple-linear-regression
* https://scikit-learn.org/stable/modules/generated/sklearn.feature_selection.RFE.html
